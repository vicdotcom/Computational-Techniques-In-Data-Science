### Question 1: Simulating a 1D Random Walk with Step Bias
A particle moves along a one-dimensional line (1D). At each time step, it can either move:
- +1 step to the right with probability p, or
- -1 step to the left with probability 1−p
Suppose p=0.85 (i.e., a bias to the right).
Instructions:
1. Write a Python program to simulate a 1D random walk with 1000 steps where: 
    -   The starting position is 0.
    - Each step has a probability of p=0.85 to move right and 1−p=0.15 to move left.
2. Plot the position vs time graph (time on x-axis, position on y-axis).
3. Run the simulation five times and overlay all five random walks in the same graph.
4. Calculate and interpret: 
    - The final position of the particle after 1000 steps.
    - The mean and standard deviation of the final position across five simulations.


```python
## probability of moving right
prob = 0.85

## No. of steps
num_steps= 1000

## Empty list that will save the final position of each random walk particle....
# ...which will be used to compute mean and standard deviation across the five random walks
final_positions= []

## Simulating 5 times
for s in range(5):
    # Size of individual steps (State space)
    steps = np.random.choice([-1, 1], size=num_steps, p=[1-prob, prob])

    # Random walk (Cumulative sum of steps)
    position = np.zeros(num_steps + 1) # Initialising the position array with zeros, plus one since the random walk has 1000 steps and we have added an initial zero to the array
    position[0]= 0
    position[1:]= np.cumsum(steps) # Cumulative sum of steps from the second element (since the first element starts with 0)

    # Final position of each random walk particle
    print(f'Random Walk {s+1} Final Position: {position[-1]}')

    # Appending the final positions of each random walk particle to the list
    final_positions.append(position[-1])


    # Plotting
    plt.plot(position)
    plt.axhline(y=position[0], color='red', linestyle='--')

plt.xlabel('Steps')
plt.ylabel('Cumulative Position')
plt.title(f'1D Unequal probability Random Walks')
plt.grid()

# Mean and standard deviation of the final positions of each random walk particle
print()
print(f'Final Positions Mean: {np.mean(final_positions)}')
print(f'Final Positions Standard Deviation: {np.std(final_positions):.2f}')
```

Due to the higher probability of moving rightwards with +1 step. The cumulative sum of steps at the 1000th step (the final position) is therefore much higher.

On average, the final position quantities for each random walk are significantly higher than the initial position and each are close to each other with minimal deviations between them.

### Question 2: Comparing 1D Random Walks with and without Drift
Suppose two particles perform 1D random walks starting from position 0:
- Particle A: Moves with a drift, i.e., p=0.7 (70% chance to move right).
- Particle B: Moves without drift, i.e., p=0.5(equal probability both sides).

Instructions:
1. Write a Python program to simulate 1000 steps for each particle.
2. Plot both random walks on the same graph with: 
    - Time on the x-axis.
    - Position on the y-axis.
    - Different colors for each particle.
3. Calculate and display:
    - The mean and standard deviation of the final position after 1000 steps.
4. Interpret your answer

```python
## Probabilty of moving to the right
prob= 0.7

## No. of steps
num_steps= 1000

## Particle A (70% probability of moving to the right)  
steps_A= np.random.choice([-1, 1], p= [1-prob, prob], size= num_steps) # Steps size
position_A= np.cumsum(steps_A) # Cumulative sum of steps- random walk

## Particle B (Equal probability of moving to the right or left)
steps_B= np.random.choice([-1, 1], p= [0.5, 0.5], size= num_steps)
position_B= np.cumsum(steps_B)

## Plotting
plt.plot(position_A, label= 'A')
plt.plot(position_B, label= 'B')
plt.axhline(y= 0, color= 'red', linestyle= '--', label= 'Zero line')

plt.legend()
plt.xlabel('Steps')
plt.ylabel('Cumulative Position')
plt.title('A random walk with drift (A) vs with no drift (B)')

## Mean and std deviation of the final positions
# Random Walk A
print('Random Walk A')
print(f'Final Position Mean: {np.mean(position_A)}')
print(f'Final Position Std Dev: {np.std(position_A):.2f}')
print()
# Random Walk B
print('Random Walk B')
print(f'Final Position Mean: {np.mean(position_B)}')
print(f'Final Position Std Dev: {np.std(position_B):.2f}')
```

**Mean**

Random walk A has a higher mean of 223.592 indicating that it has a positive drift (a tendency for the walk to move in a specific direction (upward in this case) over time). This is a determistic trend.

Random walk B has a lower mean of -7.214 which is much closer to the zero line. This indicates no drift or rather no tendency for the walk to move in a specific direction over time.

**Std Deviation**

Random Walk A has a much larger standard deviation (118.85) compared to Random Walk B (15.38), reflecting the greater variability introduced by the drift.

### Question 3: Simulating a 2D Random Walk (Unbiased)
A mosquito trapped in a square grid moves randomly:
- Up, Down, Left, or Right with equal probability (25%) in each direction.
- The mosquito starts at coordinate (0,0).

Instructions:
1. Write a Python program to simulate a 2D random walk for 500 steps.
2. Plot the path of the mosquito (X vs Y) using a scatter plot or line plot.
3. Calculate and display:
    - The final position after 500 steps.
    - The total distance from the origin after 500 steps.
4. Run the simulation 10 times and calculate: 
    - The average distance from the origin after 500 steps.
    - The standard deviation of the distance.
5. Interpret your answer

```python
## Number of steps
num_steps = 500

## Equal probability of movement in four directions (25% each)
probs = [0.25, 0.25, 0.25, 0.25]  # Left, Right, Up, Down

## List to store final positions and steps taken
final_positions = []
final_steps = []

## Colors for different random walks
colors = ['lime', 'pink', 'blue', 'indigo', 'orange', 'violet', 'cyan', 'magenta', 'yellow', 'brown']

## Initialize plot
plt.figure(figsize=(13, 7))

## Simulate 10 random walks
for s in range(10):
    # Random steps for X (left, right) and Y (up, down)
    x_steps = np.random.choice([-1, 1, 0, 0], size=num_steps, p=probs)  # Left (-1) or Right (+1), No movement (0)
    y_steps = np.random.choice([0, 0, -1, 1], size=num_steps, p=probs)  # Up (+1) or Down (-1), No movement (0)

    # Compute cumulative positions (Random Walks). Must start from 0
    x_position = np.zeros(num_steps + 1) # Intialize the array with zeros '+1' is to account for the initial zero
    x_position[1:] = np.cumsum(x_steps) # Add the cumulative sum of steps taken

    y_position = np.zeros(num_steps + 1)
    y_position[1:] = np.cumsum(y_steps)

    # Plot the random walk
    plt.plot(x_position, y_position, marker='.', linestyle='--', color=colors[s], alpha=0.8)
    plt.scatter(x_position[0], y_position[0], color=colors[s], marker='o', zorder=3, edgecolors='lime')  # Start point, marker has edgecolor lime (green)
    plt.scatter(x_position[-1], y_position[-1], color=colors[s], marker='o', zorder=3, edgecolors='red')  # End point, marker has edgecolor red

    # Store final step taken and final cumulative position (the total distance from the origin)
    final_steps.append(np.array([x_steps[-1], y_steps[-1]]))
    final_positions.append(np.array([x_position[-1], y_position[-1]]))

    # Print results
    print(f'Random Walk {s+1}')
    print(f'Final Step Taken [x,y]: {final_steps[s]}')
    print(f'Total Distance from Origin: {final_positions[s]}')
    print(f'Avg Distance Taken x-axis: {np.mean(x_position):.2f}')
    print(f'Avg Distance Taken y-axis: {np.mean(y_position):.2f}')
    print(f'Std. Deviation of the distance taken (x-axis): {np.std(x_position):.2f}')
    print(f'Std. Deviation of the distance taken (y-axis): {np.std(y_position):.2f}')
    print()

## Plot labels
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.title('10 Symmetrical 2D Random Walks')
plt.grid()
```

### Question 4: Comparing 2D Random Walks with No Bias
A person walks randomly in a 2D grid but with a slight bias towards the East (right). In each step: 
    - Move East: 40% probability
    - Move West: 20% probability
    - Move North: 20% probability
    - Move South: 20% probability

The person starts at (0,0).
1. Write a Python program to simulate: 
    - 500 steps for the biased random walk.
    - 500 steps for an unbiased random walk (equal probability).
2. Plot both paths on the same graph with: 
    - Different colors for each walk.
    - Scatter plot showing the final position.
3. Calculate and display:
    - The final position after 500 steps.
    - The total distance from the origin for both walks.
4. Run the simulation 10 times and compute: 
    - The average distance from the origin for both biased and unbiased walks.
    - The standard deviation of the distance.
5. Interpretation:
    - Why does the biased random walk drift to the east?
    - How does drift affect the standard deviation of the final position?
    - What real-world phenomena could this simulation represent (e.g., wind drift, 
ocean currents)

```python
## Probabilities of movement
probs = [0.2, 0.4, 0.2, 0.2]  # West, East, North, South

## List to store final positions and steps taken
final_positions = []
final_steps = []

## Colors for different random walks
colors = ['lime', 'pink', 'blue', 'indigo', 'orange', 'violet', 'cyan', 'magenta', 'yellow', 'brown']

## Initialize plot
plt.figure(figsize=(13, 7))

## Simulate 10 random walks
for s in range(10):
    # Random steps for X (left, right) and Y (up, down)
    x_steps = np.random.choice([-1, 1, 0, 0], size=num_steps, p=probs)  # Left (-1) or Right (+1), No movement (0)
    y_steps = np.random.choice([0, 0, -1, 1], size=num_steps, p=probs)  # Up (+1) or Down (-1), No movement (0)

    # Compute cumulative positions (Random Walks). Must start from 0
    x_position = np.zeros(num_steps + 1) # Intialize the array with zeros '+1' is to account for the initial zero
    x_position[1:] = np.cumsum(x_steps) # Add the cumulative sum of steps taken

    y_position = np.zeros(num_steps + 1)
    y_position[1:] = np.cumsum(y_steps)

    # Plot the random walk
    plt.plot(x_position, y_position, marker='.', linestyle='--', color=colors[s], alpha=0.8)
    plt.scatter(x_position[0], y_position[0], color=colors[s], marker='o', zorder=3, edgecolors='lime')  # Start point, marker has edgecolor lime (green)
    plt.scatter(x_position[-1], y_position[-1], color=colors[s], marker='o', zorder=3, edgecolors='red')  # End point, marker has edgecolor red (red)

    # Store final step taken and final cumulative position (the total distance from the origin)
    final_steps.append(np.array([x_steps[-1], y_steps[-1]]))
    final_positions.append(np.array([x_position[-1], y_position[-1]]))

    # Print results
    print(f'Random Walk {s+1}')
    print(f'Final Step Taken [x,y]: {final_steps[s]}')
    print(f'Total Distance from Origin: {final_positions[s]}')
    print(f'Avg Distance Taken x-axis: {np.mean(x_position):.2f}')
    print(f'Avg Distance Taken y-axis: {np.mean(y_position):.2f}')
    print(f'Std. Deviation of the distance taken (x-axis): {np.std(x_position):.2f}')
    print(f'Std. Deviation of the distance taken (y-axis): {np.std(y_position):.2f}')
    print()

## Plot labels
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.title('10 Symmetrical 2D Random Walks')
plt.grid()
```

**Why does the biased random walk drift to the east?**
- The biased random walk drifts to the east because there is a higher probability (40%) of moving east compared to the other directions (20% each for west, north, and south). Over many steps, this imbalance in probabilities causes the walker to move more frequently to the east, resulting in a net drift in that direction.

**How does drift affect the standard deviation of the final position?**
- The drift reduces the standard deviation of the final position compared to an unbiased random walk. This is because the drift introduces a predictable trend, reducing the overall variability in the walker's position.

**What real-world phenomena could this simulation represent?**
 - Ocean Currents: A boat or marine animal drifting in an ocean current that has a predominant direction
 - Particle Movement in a Fluid: Particles moving in a fluid with a directional flow, such as pollutants in a river.
