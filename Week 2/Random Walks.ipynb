```python
## Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

## 1D Random Walk
### Symmetric
Where a quantity moves left or right along a single axis with equal probability.

Or rather up and down along the y axis.

```python
## Number of steps
no_steps= 1000

## Steps size
 # This is the state space of the random walk process.
 # Individual steps will be either +1 or -1 with equal probability (0.5/0.5)
steps= np.random.choice([-1,1], size= no_steps, p= [0.5, 0.5])
 # p= [0.5, 0.5] means equal probability of choosing either +1 or -1 respectively
 # Does not need to be specified (its the default)

## Cummulative sum of such steps produces a random walk process
position= np.cumsum(steps)

## Plotting
plt.figure(figsize= (10,5))
plt.plot(position, label= 'Random Walk Line') 
plt.axhline(y= position[0], color= 'red', linestyle= '--', label= 'Origin') 

plt.legend()
plt.xlabel('Steps')
plt.ylabel('Cummulative Position')
plt.title('Simple Symmetrical 1D Random Walk')
plt.grid()
```

```python
np.zeros(1000)
```

So a random walk is a type of stochastic process where the current observation is equal to the previous observation plus some random change as shown by the summation of each step using `np.cumsum()`.

Mathematically, a random walk is written as:
$$X_t = X_{t-1} + w_t$$
where: 
- $X_t$ is the current observation at time $t$
- $X_{t-1}$ is the previous observation at time $t-1$
- $w_t$ is a random variable that often follows a normal distribution with mean $0$ and standard deviation $\sigma$ (Gaussian white noise)

This means that the differences between observations should be totally random, producing white noise that has no discernible pattern.
$$w_t= X_t - X_{t-1}$$

```python
## Differencing 
white = np.diff(position)
 # This is the same as the 'steps' variable
 # It's just good to know how to difference data

## Plotting the differenced data
plt.figure(figsize=(10,5))
plt.plot(white)
plt.axhline(y=0, color='red', linestyle='--', label= 'Zero Line')

plt.xlabel('Steps')
plt.ylabel('Position')
plt.title('Differenced Data/Individual Steps Plot')
plt.legend()
plt.grid()

```

### With Unequal Probability

```python
## Probability of the random walk going up
prob_up= 0.7

## State space
steps2= np.random.choice([-1,1], size= no_steps, p= [1-prob_up, prob_up])

## Random Walk
position2= np.cumsum(steps2)

## Plotting
plt.Figure(figsize= (10,5))
plt.plot(position2, label= 'Random Walk Line')
plt.axhline(y= position2[0], color= 'red', linestyle= '--', label= 'Origin line' )

plt.legend()
plt.xlabel('Steps')
plt.ylabel('Cummulative Position')
plt.title('Unequal Probability 1D Random Walk')
plt.grid()
```

Since the random walk has drift, the difference between its observations will have a clear trend.

```python
## Plotting the differences between the observations of the unequal probability random variables
plt.figure(figsize=(10,10))
plt.plot(steps2)
plt.axhline(y=0, color='red', linestyle='--', label= 'Zero Line')

plt.legend()
plt.xlabel('Steps')
plt.ylabel('Position')
plt.title('Individual Steps Plot')
```

## 2D Random Walk
Allows movement in 2 perpendicular directions (i.e- left and right, up and down) along 2 axes.
### Symmetrical


```python
## Individual steps (state space)
x_steps= np.random.choice([-1,1], size= no_steps, p= [0.5, 0.5]) # left, right
y_steps= np.random.choice([-1,1], size= no_steps, p= [0.5, 0.5]) # up, down

## Random Walks
x_position= np.cumsum(x_steps)
y_position= np.cumsum(y_steps)

## Plotting
plt.Figure(figsize= (10,6))
plt.plot(x_position, y_position, marker= '.',linestyle= '--', alpha= 0.8) # plotting the random walk movement
plt.scatter(x_position[0], y_position[0], color= 'lime', zorder= 3, label= 'Start') # plotting the starting point
plt.scatter(x_position[-1], y_position[-1], color= 'red', zorder= 3, label= 'End') # plotting the ending point

plt.legend()
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.title('Symmetrical 3D Random Walk')
plt.grid()
```

### Unequal Probability

```python
## No of steps
num_steps= 100

## Step probabilities
x_prob= 0.7 # probability of going right
y_prob= 0.7 # probability of going up

## Individual steps (state space)
x_steps= np.random.choice([-1,1], size= num_steps, p= [1-x_prob, x_prob]) # left, right
y_steps= np.random.choice([-1,1], size= num_steps, p= [1-y_prob, y_prob]) # up, down

## Random Walks
x_position= np.cumsum(x_steps)
y_position= np.cumsum(y_steps)

## Plotting
plt.Figure(figsize= (10,6))
plt.plot(x_position, y_position, marker= '.',linestyle= '--', alpha= 0.8) # plotting the random walk movement
plt.scatter(x_position[0], y_position[0], color= 'lime', zorder= 3, label= 'Start') # plotting the starting point
plt.scatter(x_position[-1], y_position[-1], color= 'red', zorder= 3, label= 'End') # plotting the ending point

plt.legend()
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.title('Unequal Probability 2D Random Walk')
plt.grid()
```
